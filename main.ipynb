```python
import pandas as pd

df = pd.read_parquet('Chatbot Arena Conversations Train.parquet')

df
```

```python
# iterate through rows

import ollama
from tqdm import tqdm

model_as = []
model_bs = []
winners = []
user_prompts = []
embeddings = []

for index, row in tqdm(df.iterrows(), total=len(df), desc="Processing rows"):
    model_as.append(row['model_a'])
    model_bs.append(row['model_b'])
    winners.append(row['winner'])
    user_prompt = (row['conversation_a'][0]['content'])
    user_prompts.append(user_prompt)
    embed = ollama.embeddings(
        model="mxbai-embed-large", 
        prompt=user_prompt
    )
    embeddings.append(embed)

df = pd.DataFrame({
    'model_a': model_as,
    'model_b': model_bs,
    'winner': winners,
    'user_prompt': user_prompts,
    'embeddings': embeddings
})

df.to_csv('chatbot_arena_train.csv', index=False)
```
